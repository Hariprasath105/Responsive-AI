```python
import csv

# Initialize an empty list to store the data
a = []

# Open the CSV file and read its content
with open(r"C:\Users\user\Downloads\data.csv", 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        a.append(row)

# Print the data
print(a)

# Calculate and print the total number of training instances
print("\nThe total number of training instances are: ", len(a))

# Determine the number of attributes (excluding the target variable)
num_attribute = len(a[0]) - 1

# Print the initial hypothesis
print("\nThe initial hypothesis is: ")
hypothesis = ['0'] * num_attribute
print(hypothesis)

# Loop through the training instances to update the hypothesis
for i in range(len(a)):
    if a[i][num_attribute].lower() == 'yes':  # If the class label is 'yes'
        for j in range(num_attribute):
            # If the attribute value is unknown (0), or matches the current hypothesis, update it
            if hypothesis[j] == '0' or hypothesis[j] == a[i][j]:
                hypothesis[j] = a[i][j]
            else:
                hypothesis[j] = '?'  # If there’s a conflict, set it to '?'
   
    # Print the hypothesis after processing each instance
    print("\nThe hypothesis for the training instance {} is: \n".format(i + 1), hypothesis)

# Print the final maximally specific hypothesis after all training instances
print("\nThe Maximally specific hypothesis for the training instance is: ")
print(hypothesis)
```
